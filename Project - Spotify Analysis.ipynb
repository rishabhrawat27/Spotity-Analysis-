# Replacing Release Date as index and converting Duration_ms into Seconds

```python
dft.set_index("release_date",inplace = True)
```

```python
dft.index= pd.to_datetime(dft.index)
dft
```

```python
dft["seconds"]=dft['duration_ms'].apply(lambda x: round(x/1000))
dft
```

```python
dft.iloc[:,3]=dft.loc[:,'seconds']
```

```python
dft.drop(columns='seconds',inplace=True)
```

```python
dft.rename(columns={'duration_ms':'seconds'},inplace=True)
dft
```

# Correlation

```python
dft.corr()
```

# Correlation Map

In the map the darker colour represent negative relationship and the brighter represent the possitive
Acousticness and Energy have the Most Negative Correlation
Loudness and Energy have the Highest Possitive Correlation

```python
plt.figure(figsize=(10,10))
plt.title('Correlation Map')
dftn = dft.drop(columns=["key","mode","explicit"]).corr()
sns.heatmap(dftn,annot =True,fmt=".1g",linewidth=1)


plt.show()
```

# Taking 0.5%  Sample of All Track

```python
sample=dft.sample(int(0.005*len(dft['id'])))
```

```python
sample
```

# Regression line of Sample Data
Direct Relationship between Energy and Loudness with the slop of 0.8

```python
plt.figure(figsize=(14,10))
sns.regplot(x="energy",y="loudness",data=sample,color="red",marker="+")
```

# Regression line of Acousticness and Popularity
Negative Relationship with the slop of -0.4

```python
plt.figure(figsize=(10,7))
sns.regplot(y="acousticness",x="popularity",data=sample,color="red",marker="+")
```

```python
dft['years']=dft.index.year
```

# Histplot

```python
sns.histplot(dft['years'],discrete=True).set_title('Releases per year')
```

# Barplot

```python
plt.figure(figsize=(15,5))
plt.title("Year vs Time")
ax=sns.barplot(x='years',y='seconds',data=dft,errwidth=False)
ax.tick_params(axis='x',rotation=90)
```

# Lineplot of Year and TIme
The length of the Song in 2021 is lowest since 1970 
The lenght of Song was maximum at 1938 and minimum at around 1922

```python
plt.figure(figsize=(14,7))
plt.title("Year vs Time")
sns.set_style(style='whitegrid')
sns.lineplot(x='years',y='seconds',data=dft)
plt.xticks(rotation=60)

plt.show()
```
